1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.



Descriptive Statistics

To organize, summarize, and present the main features of a known dataset in a clear, digestible way.

 The specific data you have (sample or entire population).

Examples:
Calculating the mean (average), median, or mode of test scores for one classroom.

Creating a histogram or pie chart showing the frequency of different car colors sold.

Finding the range or standard deviation of patient blood pressures in a clinic.

Inferential Statistics

To draw conclusions, make predictions, or test hypotheses about a larger population based on a smaller sample.

 Making educated guesses about the broader population.

Examples:
Using a sample of voters to predict the outcome of a national election.

A drug company testing a new medicine on a small group to see if it's effective for all patients (using confidence intervals).

A researcher using a t-test to see if basketball players are significantly taller than the general male population.

2.  What is sampling in statistics? Explain the differences between random
and stratified sampling.

Sampling is the process of selecting a small group (sample) from a large group (population) in order to study and draw conclusions about the entire population.
It is used because studying the whole population is often time-consuming, costly, and impractical.

1. Random Sampling

In random sampling, every member of the population has an equal chance of being selected.

Example:
From a class of 100 students, 10 students are selected by lottery method or computer-generated random numbers.

2. Stratified Sampling

In stratified sampling, the population is first divided into sub-groups (strata) based on a common characteristic (like age, gender, income). Then random samples are taken from each stratum.

Example:
A college has:
60% girls
40% boys
If 10 students are selected:
6 girls are chosen randomly
4 boys are chosen randomly

3. Define mean, median, and mode. Explain why these measures of central
tendency are important.

Mean: The arithmetic average; calculated by summing all values and dividing by the count of values. It's sensitive to extreme values (outliers).

Median: The middle value in a dataset when arranged in order. If there are two middle numbers, it's their average. It's less affected by outliers than the mean.

Mode: The value that appears most often in the dataset. A dataset can have one mode, multiple modes, or no mode.

Use Mean for balanced, symmetrical data without outliers (e.g., test scores in a typical class).

Use Median for skewed data or data with outliers (e.g., income data, where a few high earners skew the mean).

Use Mode for categorical data or to find the most common occurrence (e.g., favorite color, most common shoe size).

4. Explain skewness and kurtosis. What does a positive skew imply about the
data?

Skewness

Zero Skew: Symmetrical; left and right sides mirror each other (e.g., a perfect bell curve).
Positive (Right) Skew: Tail points right; most data on the left; Mean > Median > Mode.
Negative (Left) Skew: Tail points left; most data on the right; Mean < Median < Mode.

Kurtosis

Measures how heavy the tails are or how sharp the peak is relative to a normal distribution.
Leptokurtic (High Kurtosis): Heavier tails, sharper peak, more outliers (kurtosis > 3).
Platykurtic (Low Kurtosis): Lighter tails, flatter peak, fewer outliers (kurtosis < 3).
Mesokurtic: Similar to normal distribution (kurtosis = 3).

5.: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

```python
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.multimode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:

list_x = [10, 20, 30, 40, 50]

list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

x = np.array([10, 20, 30, 40, 50])
y = np.array([15, 25, 35, 45, 60])

covariance = np.cov(x, y)[0][1]
correlation = np.corrcoef(x, y)[0][1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```

8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.

● Explain how you would use covariance and correlation to explore this
relationship.

● Write Python code to compute the correlation between the two lists:

advertising_spend = [200, 250, 300, 400, 500]
  
daily_sales = [2200, 2450, 2750, 3200, 4000]

```python
import numpy as np

advertising_spend = np.array([200, 250, 300, 400, 500])
daily_sales = np.array([2200, 2450, 2750, 3200, 4000])

covariance = np.cov(advertising_spend, daily_sales)[0][1]
correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```

9. Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.

● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.

● Write Python code to create a histogram using Matplotlib for the survey data:

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

```python
import matplotlib.pyplot as plt
import numpy as np

survey_scores = [7,8,5,9,6,7,8,9,10,4,7,6,9,8,7]

mean = np.mean(survey_scores)
std_dev = np.std(survey_scores)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

plt.hist(survey_scores, bins=5)
plt.title("Customer Satisfaction Survey")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```
